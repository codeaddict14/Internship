```python
#importing the required libraries
import selenium 
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
```

```python
#connecting with the drivers
driver=webdriver.Chrome(r"C:\Users\MICROSOFT\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
#connecting with the website extracting information from flipkart for laptop
driver.get("https://www.flipkart.com/dell-inspiron-core-i5-10th-gen-8-gb-512-gb-ssd-windows-10-home-ins-5408-thin-light-laptop/product-reviews/itmc78f94d1da92c?pid=COMFUZ3JZ6VQZN9Y&page=1")
```

```python
ratingflip=[] #extracting rating given by users and appending it in list.

def rating():
    rati=driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq' or @class='_3LWZlK _32lA32 _1BLPMq']")
    for i in rati:
        ratingflip.append(i.text)
    return(ratingflip)
```

```python
review2=[] #extracting reviews and appending it in list.
def review():
    rev=driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
    for i in rev:
        review2.append(i.text.replace("\n"," "))
    return(review2)
```

```python
i=0
while(i<6):
    try:
        driver.find_elements_by_xpath("//a[@class='_1LKTO3']")[-1].click() #(clicking on the next button to reach last)
        rating()
        review()
    except:
        pass
        
    i=i+1
```

```python
#contecting with the website flipkart and extracting reviews of iphone 11.
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/product-reviews/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace=FLIPKART")
```

```python
counter = 0
data = []

while(counter<1000):
    reviews_deck = driver.find_elements(By.XPATH,"//div[@class='col _2wzgFH K0kLPL']")    
    for review in reviews_deck:
        
        if(counter>1000):
            break
        else:
            counter+=1
        try:
            review.find_element(By.XPATH,".//div[@class='row']//span[@class='_1BWGvX']").click()
        except:
            pass
        finally:
            rating, review_desc = review.find_elements(By.XPATH,".//div[@class='row']")[:2]
            rating_stars, rating_desc = rating.text[0],rating.text[1:]
        
        rating_stars_ = rating_stars.strip()
        rating_desc_ = rating_desc.strip()
        review_desc_ = review_desc.text.strip()
        data.append([rating_stars_, rating_desc_, review_desc_])
    try:
        driver.find_element(By.XPATH,"//a[@class='_1LKTO3'][2]/span").click()
        time.sleep(1)
    except:
        driver.find_element(By.XPATH,"//a[@class='_1LKTO3']/span").click()
        time.sleep(2)
```

```python
columns = ['Rating', 'Rating Summary', 'Review']
df = pd.DataFrame(data, columns=columns)
```
