```python
#importing the required libraries
import selenium 
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
from selenium.webdriver.common.by import By
```

```python
#connecting with the drivers
driver=webdriver.Chrome(r"C:\Users\MICROSOFT\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
#extracting data from make my trip.com
driver.get("https://www.makemytrip.com/flight/search?itinerary=DEL-BLR-28/01/2022&tripType=O&paxType=A-1_C-0_I-0&intl=false&cabinClass=E&ccde=IN&lang=eng&cmp=SEM|D|DF|G|Brand|B_M_Makemytrip_Variants|Brand-Variants-Exact|RSA|Regular|V2|529579610088")
```

```python
flightname=[] #Airline name
name=driver.find_elements_by_xpath("//span[@class='boldFont blackText airlineName']")
for i in name:
    flightname.append(i.text.strip(" "))
```

```python
flightname[:5]
```

```python
len(flightname)
```

```python
depttime=[] #deprature time of the flight
time=driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoLeft']/div/p")
for i in time:
    depttime.append(i.text)
```

```python
departures=depttime[::2]
departures[:5]
```

```python
source=depttime[1::2] #source from where the flight take off
source[:5]
```

```python
arrivaltime=[] #time at which the flight lands
atime=driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div/p")
for i in atime:
    arrivaltime.append(i.text)
```

```python
arrivals=arrivaltime[::2]
arrivals[:5]
```

```python
destination=arrivaltime[1::2] #destination of the flight
destination[:5]
```

```python
stopinfo=[]  #duration of the flight
stop=driver.find_elements_by_xpath("//div[@class='stop-info flexOne']/p")
for i in stop:
    stopinfo.append(i.text)
```

```python
stopinfo[:5]
```

```python
layover=[]  #number of stops
lay=driver.find_elements_by_xpath("//p[@class='flightsLayoverInfo']")
for i in lay:
    layover.append(i.text)
    
```

```python
layover[:5]
```

```python
price=[] #price of the flight
pr=driver.find_elements_by_xpath("//div[@class='textRight flexOne']/p")
for i in pr:
    price.append(i.text)
```

```python
price[:5]
```

```python
job=pd.DataFrame({})  #creating the dataframe and saving it in different titles.
job['Airline']=flightname
job['Source']=source
job['Destination']=destination
job['Departure']=departures
job['Arrival']=arrivals
job['Duration']=stopinfo
job['Stops']=layover
job['Price']=price
job
```

```python
job_new=job.assign(Date_of_journey='28 Jan 2022')
job_new
```
