5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used


```python
#loading the server
driver=webdriver.Chrome(r"C:\Users\MICROSOFT\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#loading the web page
driver.get("https://github.com/")
```

```python
#clicking on explore
explore=driver.find_element_by_xpath("//nav[@class='d-flex flex-column flex-md-row flex-self-stretch flex-md-self-auto']/a[4]")
explore.click()
```

```python
#clicking on trending
trending=driver.find_element_by_xpath("//div[@class='d-flex flex-wrap flex-items-center flex-justify-center flex-md-justify-start text-center text-md-left']/a[3]")
trending.click()
```

```python
title=[] #extracting on the title 
tit=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
for i in tit:
    title.append(i.text)
len(title)
```

```python
desp=[]   #extracting the description
des=driver.find_elements_by_xpath("//p[@class='col-9 color-fg-muted my-1 pr-4']")
for i in des:
    desp.append(i.text)
len(desp)
```

```python
count=[]   #extracting the contibutors count
cou=driver.find_elements_by_xpath("//div[@class='f6 color-fg-muted mt-2']/a[2]")
for i in cou:
    count.append(i.text)
len(count)
```

```python
urls=[] #extracting the urls of each repository then finding the laguage of each repository
url=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in url:
    urls.append(i.get_attribute("href"))
```

```python
lan=[]  #iterating through the urls finding the languages
for i in urls:
    driver.get(i)
    try:
        la=driver.find_element_by_xpath("//a[@class='d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3']/span")
        lan.append(la.text)
    except:
        lan.append("--")
len(lan)
```

```python
job6=pd.DataFrame({})
job6['Repo Title']=title
job6['Repo Description']=desp
job6['Contributor count']=count
job6['Language']=lan
job6
```

```python
name=[] #extracting the name and saving in the empty list
na=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in na:
    name.append(i.text)
name[:5]
```

```python
year=[]  #extracting the year and saving it in the year list
na=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")
for i in na:
    year.append(i.text)
year[:5]
```

```python
data1=[] #extracting the time and genere
columns=['time','genere']
items=driver.find_elements_by_xpath("//p[@class='text-muted text-small']")
for item in items:
    try:
        time=item.find_element_by_class_name("runtime").text.strip()
    except:
        pass
    try:
        genre=item.find_element_by_class_name("genre").text.strip()
    except:
        pass
    data1.append([time,genre])
```

```python
data2=data1[::3]
data2[:5]
```

```python
rating=[]  #extracting the rating
ra=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
for i in ra:
    rating.append(i.text)
len(rating)
```

```python
votes=[] #extracting the voting
vo=driver.find_elements_by_name("nv")
for i in vo:
    votes.append(i.text)
votes[:5]
```

```python
job2=pd.DataFrame({})  #saving in the dataframe format
job2['Name']=name
job2['Year Span']=year
job2['Time and genere']=data2
job2['Rating']=rating
job2['Votes']=votes
job2
```
