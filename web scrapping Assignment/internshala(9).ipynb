### Write a python program to scrape fresher job listings from ‘https://internshala.com/’. 
It should include job title, company name, CTC, and apply date.


```python
import bs4
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://internshala.com/fresher-jobs")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
title=soup.find_all("div",class_="heading_4_5 profile")
title
```

```python
titles=[]
for i in title:
    titles.append(i.text.strip().replace("\n"," "))
```

```python
titles
```

```python
company=soup.find_all("div",class_="heading_6 company_name")
company
```

```python
company1=[]
for i in company:
    company1.append(i.text.strip().replace("\n"," "))
```

```python
company1
```

```python
ctc=soup.find_all("div",class_="item_body")
ctc
```

```python
fctc=[]
for i in ctc:
    fctc.append(i.text.strip().replace("\n"," "))
```

```python
fctc
```

```python
fctcf=fctc[1::3]
fctcf
```

```python
applydate=fctc[2::3]
applydate
```

```python
import pandas as pd
job=pd.DataFrame({})
job["Title"]=titles
job["Company"]=company1
job["CTC"]=fctcf
job["Apply Date"]=applydate
```
